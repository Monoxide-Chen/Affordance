```python
import numpy as np
import pickle
import pandas as pd

# Load .npy file
data = np.load('/storage_fast/ycli/3d_affordance/run_analysis/train_IOU.npy')

# Load shape_ids list
with open('/storage_fast/ycli/3d_affordance/run_analysis/train_shapeid_list.pkl', 'rb') as f:
    shape_ids = pickle.load(f)

# Create a DataFrame from the data array and shape_ids list
df = pd.DataFrame(data, columns=["object_class", "affordance_class", "iou"])
df["shape_id"] = shape_ids
# df.head()
# Convert object_class and affordance_class columns to integer data type
df["object_class"] = df["object_class"].astype(int)
df["affordance_class"] = df["affordance_class"].astype(int)

# Reorder the columns as requested
df = df[["shape_id", "object_class", "affordance_class", "iou"]]

# Sort DataFrame by shape_id
df = df.sort_values(by="shape_id")

# Calculate the average iou for each shape_id
average_iou = df.groupby("shape_id")["iou"].mean()
```

```python
# Merge the average_iou Series with the original DataFrame
average_iou_df = average_iou.reset_index().rename(columns={"iou": "average_iou"})
merged_df = df.merge(average_iou_df, on="shape_id")
merged_df
# # Group the merged DataFrame by "object_class" and aggregate the average IoU for each group
# grouped_by_object_class = merged_df.groupby("object_class")["average_iou"].agg(["mean", "count"])

# # Reset the index to make "object_class" a column
# grouped_by_object_class = grouped_by_object_class.reset_index()

# grouped_by_object_class
```
